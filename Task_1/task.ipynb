# Part B: Data Cleaning & Transformation 

```python
# handling the numberic missing Values
# Checking the percentage of missing values in each numeric column of transactionsDf
print("Percentage of missing values in each column of transactionsDf:- \n")
for column in transactionsDf.columns:
    print(column + ": " + str(round(transactionsDf[column].isnull().mean()*100,2)) + " %" )
print("\n")

# Checking the percentage of missing values in each numeric column of customersDf
print("Percentage of missing values in each column of customersDf:- \n")
for column in customersDf.columns:
    print(column + ": " + str(round(customersDf[column].isnull().mean()*100,2)) + " %" )    
print("\n")

# checking the percentage of missing values in each numeric column of productsDf
print("Percentage of missing values in each column of productsDf:- \n")
for column in productsDf.columns:
    print(column + ": " + str(round(productsDf[column].isnull().mean()*100,2)) + " %" ) 
print("\n")
# As the percentage of missing values in all the columns is <5% Dropping them is an option, but
# it is always better to fill them with mean(for numeric values) or mode(for categorial values / objects) 
# to avoid unnecessary data loss
```

```python
# Filling missing numeric values in transactionsDf
for column in transactionsDf.columns:
    if transactionsDf[column].dtype in ['int64', 'float64']:
        mean_val=transactionsDf[column].mean()
        transactionsDf[column]=transactionsDf[column].fillna(mean_val)

# Filling missing numeric values in customersDf
for column in customersDf.columns:
    if customersDf[column].dtype in ['int64', 'float64']:
        mean_val=customersDf[column].mean()
        customersDf[column]=customersDf[column].fillna(mean_val)

# note:- 1. Since ProductsDf has no numeric missing values, no need for handling
#        2. While using the fillna() method with inplace=True, it leads to chained-assignment warnings.
#           Therefore, we assign the filled series back to the DataFrame column to avoid such warnings
```

```python
# Filling missing categorial values in transactionsdf
transactionsDf['payment_method'] = transactionsDf['payment_method'].fillna('Unknown')

# Filling missing categorial values in customersdf
customersDf['email'] = customersDf['email'].fillna('Unknown')

# Filling missing categorial values in productsdf
productsDf['category'] = productsDf['category'].fillna('Unknown')
```

```python
# Verifying missing values after filling values
print("Missing values after inputing values:- \n")
print("TransactionsDf Missing Values:- \n", transactionsDf.isnull().sum(), "\n")
print("CustomersDf Missing Values:- \n", customersDf.isnull().sum(), "\n")
print("ProductsDf Missing Values:- \n", productsDf.isnull().sum(), "\n")
```

```python
# Merging DataFrames
transactions_newDf =transactionsDf
customers_newDf = customersDf
products_newDf = productsDf
print(transactions_newDf.shape)
print(customers_newDf.shape)
print(products_newDf.shape)
# Merged DataFrames are named as transactions_newDf, customers_newDf and products_newDf
completeDf = transactions_newDf.merge(customers_newDf, on='customer_id', how='left').merge(products_newDf, on='product_id', how='left')
print("The completeDf shape after merging all three DataFrames is:- ", completeDf.shape)
print("Datatypes of completeDf:- \n", completeDf.dtypes)
print("Missing values in completeDf:- \n", completeDf.isnull().sum())
# In here total 10 columns are added after merging all three DataFrames and no of rows remain same as transactionsDf
#also there are no missing values in completeDf
# The left join is used to ensure all transactions are retained even if customer or product details are missing
```

```python
# Creating revenue column revenur = quantity * price
completeDf['revenue']=completeDf['quantity']*completeDf['price']

# Creating profit margin column profit_margin = (price - cost_price) / price × 100
completeDf['profit_margin'] = (completeDf['price']-completeDf['cost_price']) / completeDf['price']*100

print(completeDf['revenue'].head(2))
print(completeDf['profit_margin'].head(2))
```

# Part C: Aggregation & Analysis 

```python
# creating empty DataFrames for Part C: Aggregation & Analysis
c1_df = pd.DataFrame()            # Df for Customer Metrics
c2_df = pd.DataFrame()            # Df for Time-based Analysis  
#c3_df = pd.DataFrame()             Df for Product Performance
#c4_df = pd.DataFrame()             Df for Customer Segmentation
#completeDf.columns
```

```python
# Performing Part C1: Customer Metrics

# Creating total_revenue column
c1_df['total_revenue'] = completeDf.groupby('customer_id')['revenue'].sum()

# Creating total_transactions column
c1_df['total_transactions'] = completeDf.groupby('customer_id')['transaction_id'].nunique()         # nunique() method retuns the count of unique values

# Creating average_transaction_value column
c1_df['average_transaction_value'] = c1_df['total_revenue']/c1_df['total_transactions']

# Creating most_frequently_purchased_category column
c1_df['most_freuqntly_purchased_category'] = completeDf.groupby('customer_id')['category'].agg(lambda x: x.mode()[0])
# here we are using agg() method with lambda function to get the mode of category for each customer_id
# mode()[0] is used to get the first mode in case of multiple modes, as mode() returns a Series
c1_df = c1_df.sort_values('total_revenue',ascending = False)
# Displaying c1_df
c1_df
```

```python
# performing Part C2: Time-based Analysis

# Creating revenue column revenue_per_month column
c2_df['revenue_per_month'] = completeDf.groupby('month')['revenue'].sum()

# Creating number_of_unique_customers_per_month comlumn
c2_df['number_of_unique_customers_per_month'] = completeDf.groupby('month')['customer_id'].nunique()

# Creating average_order_value_per_month column
c2_df['average_order_value_per_month'] = c2_df['revenue_per_month'] / c2_df['number_of_unique_customers_per_month']

# Creating month_over_month_revenue_growth column (formula:-current_month_revenue - previous_month_revenue) / previous_month_revenue × 100
c2_df['month_over_month_revenue_growth'] = (c2_df['revenue_per_month']-c2_df['revenue_per_month'].shift(1))*100 / c2_df['revenue_per_month'].shift(1)
#here .shift(1) function shifts the values by 1 position to get previous month revenue

# Displaying c2_df
c2_df
```

```python
# Performing Part C3: Product Performance

# Finding top 10 products by revenue, total quantity sold, and profit margin and storing them in respective lists
top_10_products_by_revenue = (completeDf.groupby('product_name')['revenue'].sum().sort_values(ascending=False).head(10).index.tolist())
top_10_products_by_total_quantity_sold = (completeDf.groupby('product_name')['quantity'].sum().sort_values(ascending=False).head(10).index.tolist())
top_10_products_by_profit_margin = (completeDf.groupby('product_name')['profit_margin'].mean().sort_values(ascending=False).head(10).index.tolist())

# Creating c3_df DataFrame with the top 10 products lists in dictionary format
c3_df = pd.DataFrame({
    'top_10_products_by_revenue': top_10_products_by_revenue,
    'top_10_products_by_total_quantity_sold': top_10_products_by_total_quantity_sold,
    'top_10_products_by_profit_margin': top_10_products_by_profit_margin
})

# Displaying c3_df
c3_df
```

```python
# Performing Part C4: Customer Segmentation
# Recency analysis
max_date = completeDf['timestamp'].max()
last_transaction_date = completeDf.groupby('customer_id')['timestamp'].max()
recency = (max_date - last_transaction_date).dt.days

# Frequency analysis
total_number_of_transactions = completeDf.groupby('customer_id')['transaction_id'].nunique()

# Monetary analysis
total_revenue = completeDf.groupby('customer_id')['revenue'].sum()

# Creating c4_df DataFrame with recency, frequency, and monetary values
c4_df = pd.DataFrame({
    'recency': recency,
    'frequency': total_number_of_transactions,
    'monetary': total_revenue
})

# Displaying c4_df
c4_df
```

# Part D: Advanced Operations 

```python
# Identify suspicious transactions defined as:

def find_suspicious_transactions(full_data):
    """
    Identify suspicious transactions using vectorized operations.
    Returns: DataFrame with columns [transaction_id, customer_name, reason]
    """
    # Vectorized filtering of entire DataFrame for suspicious transactions
    transaction = full_data[((full_data['quantity'] > 100) & (full_data['price'] < 10)) | (full_data.groupby(['customer_id', 'hour'])['transaction_id'].transform('count') > 3)]
    
    # Determine reasons for suspicion using vectorized operations and lambda function
    reasons = [transaction.apply(lambda row: 'Quantity > 100 and Price < 10' if (row['quantity']>100 and row['price']<10) else 'More than 3 purchases in a single hour', axis=1)]
    
    # Constructing and returning the result DataFrame
    return pd.DataFrame({
        'transaction_id': transaction['transaction_id'],
        'customer_name': transaction['customer_name'],
        'reason': reasons[0]
    })

# Finding suspicious transactions
suspicious_transactions_Df=find_suspicious_transactions(completeDf).reset_index(drop=True)

# Displaying suspicious_transactions_Df
print(suspicious_transactions_Df)
```

```python
# Rolling Average Analysis
seven_day_rolling_Avg = completeDf.sort_values(['customer_id','timestamp']).set_index('timestamp').groupby('customer_id')['revenue'].rolling('7D',min_periods=1).mean().reset_index()
seven_day_rolling_Avg.head()

# the rolling function here is use to perform operaions on a dataframe or series over a period/window fo time 
# Syntax:-DataFrame/Series.rolling(window, min_periods=None, center=False, win_type=None, on=None, axis=0, closed=None)
```

```python
# Feature Engineering

# days_since_signup: Days between signup date and first transaction
completeDf['signup_date'] = pd.to_datetime(completeDf['signup_date'])
signup_Date = completeDf.groupby('customer_id')['signup_date'].min()     
transaction_dates = completeDf.groupby('customer_id')['timestamp'].min()
days_since_signup = abs((transaction_dates - signup_Date).dt.days)               # dt.days gets the difference in days

# purchase_frequency: Average days between purchases
pruchase_frequency = completeDf.groupby('customer_id')['timestamp'].apply(lambda x: x.sort_values().diff().dt.days.mean())      # diff() function calculates the difference between consecutive transaction dates after sorting them (for dates it returns Timedelta (diff btwn consecutive timestamps) object)

# category_diversity: Number of unique product categories purchased
category_diversity = completeDf.groupby('customer_id')['category'].nunique()

# preferred_payment: Most used payment method
preferred_payment = completeDf.groupby('customer_id')['payment_method'].agg(lambda x: x.mode()[0])

feature_engineering = pd.DataFrame({
    'days_since_signup': days_since_signup,
    'pruchase_frequency': pruchase_frequency,
    'category_diversity':category_diversity,
    'preferred_payment': preferred_payment
})
# Displaying first 5 rows of fearure_engineering
feature_engineering.head()
```

```python
# E2. Reusable Function
transactions_newDf['revenue'] = completeDf['revenue']       #adding the Revenue column from completeDf to transaction_newDf
def customer_lifetime_value(transactions_df, customer_id, discount_rate=0.1):
    
    """
    Calculate customer lifetime value (CLV) with time-discounting.
    
    CLV = Σ(revenue × discount_factor^month)
    where month is indexed from the first purchase
    """
    # Get first transaction date for each customer
    first_transaction = transactions_df.groupby('customer_id')['timestamp'].transform('min')

    # Calculate months since first purchase (no map used)
    transactions_df['month_index'] = (transactions_df['timestamp'].dt.to_period('M').astype(int)- first_transaction.dt.to_period('M').astype(int))
    #Note:-  dt.to_period('M') converts to monthly period (i.e. Gives monthly period), astype(int) converts period to integer for calculation
    '''        similar to 'M' there are:- 
                Frequency code	Meaning
                'D'	Day
                'M'	Month
                'Q'	Quarter
                'Y'	Year
                'H'	Hour '''
    # Compute CLV per customer
    CLV = transactions_df.groupby('customer_id').apply(lambda x: (x['revenue'] * ((1 - discount_rate) ** x['month_index'])).sum())
        # include_groups = False to avoid multi-level index and warning, this is because we are grouping by customer_id and applying a function
    return CLV

# Example usage
CLV = customer_lifetime_value(transactions_newDf, transactions_newDf['customer_id'])
print(CLV.head())
```
